Buscamos graficar $x_{\infty}$ contra $\mu$ para las siguientes ecuaciones:
$$ x_{n+1}=\mu x_n(1- x_n) \qquad x\in [0,1] \quad \mu \in [0,4]$$

Tenemos que:

```julia
# Definimos nuestros valores de m con un linspace

m = linspace(0,4,500)
```

```julia
using PyPlot
```

```julia
# CÃ³digo para graficar los puntos fijos en funciÃ³n del parÃ¡metro m.

for i in 1:length(m)
v = [] # Arreglo que contiene los puntos fijos.
    
    # Definimos un for para definir la funciÃ³n f(x) para cada valor de m.
    for j in 1:length(m) 
        xval = zeros(100) # Lista con valores de x
        xval[1] = rand() # Definimos la condiciÃ³n inicial para introducir en nuestra fÃ³rmula de recursiÃ³n
        f(x) = m[j]*x*(1-x) # FunciÃ³n de recursiÃ³n
            
            # Este for es para calcular los puntos fijos usando la funciÃ³n f(x) definida para un valor de m.
            for k in 1:length(xval)-1
                xval[k+1] = f(xval[k])
            end
            push!(v,xval[length(xval)]) # Guardamos el punto fijo el nuestro arreglo.
    end
    scatter(m,v,s=0.2) # Graficamos los puntos fijos para un valor de m dado.
    
end

ylabel("Puntos fijos")
xlabel("m")
title("Mapeo logÃ­stico de x_(n+1)=m*x_n*(1-x_n)")
grid("on")
```
