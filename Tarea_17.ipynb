## Ejercicio 2

¿Como se define una matriz simétrica?

Una matriz simétrica A es una matriz cuadrada de con n x n elementos tales que sus entradas cumplen que $a_{ij}=a_{ji}$ para toda $i,j=1,2,\dots,n$. Un ejemplo de una matriz simétrica es la siguiente:

$$A=\left(\begin{array}{cc} 3 & -1\\ -1 & 2 \end{array} \right).$$

donde $a_{12}=a_{21}=-1$.

Por definición, una matriz simétrica es igual a su traspuesta: $A=A^t$.

¿Como puede construir una matriz simétrica a partir de cualquier matriz cuadrada?

Para hacer esto, buscamos una matriz $D$ diagonal que sea semejante a una matriz cuadrada arbitraria $B$.

Definimos una matriz cuadrada B con valores arbitrarios:

```julia
B = 5*rand(10,10)
```

```julia
# Definimos una nueva matriz como la suma de la matriz arbitraria B con su traspuesta

Bs=B+B'
```

```julia
# Restando esta nueva matriz con su traspuesta

Bs - Bs'
```

Por lo tanto, podemos construir una matriz simétrica $B=A+A^t$ a partir de una matriz cuadrada arbitraria $A$.

## Ejercicio  4 

Construya el propagador $e^{A t}$ para cada tiempo sobre una malla y uselo para propagar un estado inicial $\vec x_0$ cualquiera. Utilice:
$$A=\left(\begin{array}{cc} -1 & 1\\ -1 & -1 \end{array} \right).$$
* Grafique la trayectoria en el espacio de fase.

```julia
# Escribimos nuestra matriz A

A = [-1 1;-1 -1]
```

```julia
# Buscamos los eigenvalos de la matriz A

vals=eigvals(A)
```

```julia
# Definimos una matriz diagonal cuyas entradas son los eigenvalores de A

d=diagm(vals)
```

```julia
# Definimos una matriz cuyas columnas son los eigenvectores de la matriz A.

w = eigvecs(A)
```

```julia
# Nuestro propagados exp(At) esta dado por

eA(t) = w*diagm(exp.(vals*t))*inv(w)
```

```julia
eA(3)
```

```julia
# Definimos una función de la solución a nuestra ecuación diferencial: x=(w*exp(A*t)*w^-1)*x0 donde los parámetros son el valor
# t y una condición inicial x0.

E(t,x0) = w*diagm(exp(vals*t))*inv(w)*x0
```

```julia
# Definimos nuestra lista de valores t.

t1 =  linspace(0,5,100)
```

```julia
# Llamamos el paquete PyPlot para graficar.

using PyPlot
```

```julia
# Definimos un ciclo para evaluar exp(At)) en cada valor de nuestra linspace y otro ciclo para graficar muchas curvas x contra u

# El ciclo con índice j es para graficar muchas curvas con diferentes condiciones iniciales
for j in 1:500
    
    # Definimos nuestra matriz vacía que contendrá los elementos de w*exp(At)*w^-1*x0.
    X = []
    xr = 3*[2*rand()-1; 2*rand()-1]
    
    # El ciclo con índice i es para meterlos valores en nuestra lista X. Evaluamos nuestra función en cada valor de t e introducimos este valor a nuestro arreglo vacío X.
    for i in 1:length(t1)
        e = E(t1[i], xr)
        push!(X,e)
    end
    
    # Al terminar cada ciclo, graficamos todos los valores del arreglo X.
    listx=map(x->x[1],real(X))
    listu=map(x->x[2],real(X))
    plot(listx,listu)
end
title("Espacio fase")
xlabel("x(t)")
ylabel("u(t)")
grid("on")

show()
```
